```python
import numpy as np
import pandas as pd

```

```python


df = pd.read_csv("/content/drive/MyDrive/dataset/DataSet_Animal_Center_Intakes.csv")

print("Shape of the dataset:", df.shape)

columns_with_null = df.columns[ df.isnull().any()]
print("Columns with null values:", columns_with_null)

df.drop_duplicates(inplace=True)

df.rename(columns={
    'Animal ID': 'id',
    'DateTime': 'date_time',
    'MonthYear': 'month_year',
    'Found Location': 'found_location',
    'Intake Type': 'intake_type',
    'Intake Condition': 'intake_condition',
    'Animal Type': 'animal_type',
    'Sex upon Intake': 'sex_upon_intake',
    'Age upon Intake': 'age_upon_intake',
    'Breed': 'breed'
}, inplace=True)

df.drop(columns=['month_year'], inplace=True)

df['month'] = pd.to_datetime(df['date_time']).dt.month
df['year'] = pd.to_datetime(df['date_time']).dt.year

yearly_intake = df.groupby('year')['id'].count()
print("Year-wise Animal Intake (2014 to 2021):")
print(yearly_intake)

monthly_intake = df.groupby('month')['id'].count()
print("Month-wise Animal Intake:")
print(monthly_intake)

intake_methods_count = df['intake_type'].value_counts()
print("Animal Intake Methods Count:")
print(intake_methods_count)

animal_conditions = df['intake_condition'].unique()
print("Conditions of animals taken into the shelter:")
print(animal_conditions)

animal_types_count = df['animal_type'].value_counts()
print("Types of animals taken into the shelter:")
print(animal_types_count)

top_10_breeds = df['breed'].value_counts().head(10)
print("Top 10 Breeds in Animal Shelter:")
print(top_10_breeds)

top_20_names = df['Name'].value_counts().head(20)
print("Top 20 Common Names of Animals in Animal Shelter:")
print(top_20_names)

df_outcomes = pd.read_csv("/content/drive/MyDrive/dataset/DataSet_Animal_Center_Intakes.csv")

if 'Outcome Type' in df_outcomes:
    outcome_types_count = df_outcomes['Outcome Type'].value_counts()
    print("Outcome Type-wise Count of Animals:")
    print(outcome_types_count)
else:
    print("Outcome Type data not available.")
```
